<H1>English words - English to Spanish </H1>

```python
df_expt1=pd.read_csv('expt1.csv',encoding="LATIN1")
df_expt1.head(5)
```

```python
df_expt1['Valence'] = 0.0
df_expt1.head(5)
```

```python
import nltk
from nltk.corpus import wordnet
```

```python
for index, row in df_expt1.iterrows():
    #print(row['word'], row['freq'],row['Valence'])
    synonyms = []
    for syn in wordnet.synsets(row['word']):
       for l in syn.lemmas():
            synonyms.append(l.name())
    match=list((set(df_vad["Word"]) & set(synonyms)))
    if match:
        word=match[0]
        val=df_vad[df_vad['Word'] == word]["V.Mean.Sum"]
        df_expt1.set_value(index,"Valence",val)
        #row['Valence']=df_vad[df_vad['Word'] == row['word']]["V.Mean.Sum"]
        
```
